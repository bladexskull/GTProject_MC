### Import libraries

```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Reading Data

```python
df=pd.read_excel('Hypothetical data of friends.xlsx');
```

#### Data Format

```python
df.head()
```

```python
Friend_list1=np.array([])
Friend_list2=np.array([])
```

```python
Friend_list1=df['From']
Friend_list2=df['To']
```

```python
g=nx.Graph()
```

### Building Graph by adding Edges

```python
for x in range(118):
    g.add_edge(Friend_list1[x],Friend_list2[x])
```

### Visualising Graph


```python
fig = plt.figure(figsize=(6, 6))
colors = np.linspace(0, 1, len(g.nodes))
nx.draw(g, 
        node_size=50,
        node_color=colors,
        edge_color='white')
fig.set_facecolor('black')
```

<h3> Visualising Graph with labels </h3>


```python
fig=plt.figure(figsize =(9, 9)) 
nx.draw(g, 
        node_size=110,
        node_color=colors,
        edge_color='white',with_labels=True)
fig.set_facecolor('SaddleBrown')
```

#### Degree Centrality

```python
most_influential = nx.degree_centrality(g)
```

```python
i=0;
Dict={'Name':[],'Degree_centrality':[]}
for w in sorted(most_influential, key=most_influential.get, reverse=True):
    i=i+1;
    if i<11 :
        Dict['Name'].insert(i,w)
        Dict['Degree_centrality'].append(most_influential[w])
```

```python
df_deg = pd.DataFrame(Dict)
df_deg
```

### EigenVector Centrality

```python
most_influential2=nx.eigenvector_centrality(g)
```

```python
i=0;
Dict={'Name':[],'Eigenvector_centrality':[]}
for w in sorted(most_influential2, key=most_influential2.get, reverse=True):
    if i<10 :
        Dict['Name'].insert(i,w)
        Dict['Eigenvector_centrality'].append(most_influential2[w])
        i=i+1;
```

```python
df_Eig = pd.DataFrame(Dict)
df_Eig
```

### Betweenness Centrality

```python
nx.betweenness_centrality(g)
most_influential3=nx.betweenness_centrality(g)
```

```python
i=0;
Dict={'Name':[],'Betweenness_centrality':[]}
for w in sorted(most_influential3, key=most_influential3.get, reverse=True):
    if i>=0 and i<10 :
        Dict['Name'].insert(i,w)
        Dict['Betweenness_centrality'].append(most_influential3[w])
        i=i+1;
```

```python
df_bet = pd.DataFrame(Dict)
df_bet
```

###  Visualising Degree Centrality Graph

```python
fig = plt.figure(figsize=(9, 9))

centrality = nx.degree_centrality(g)
cent_vals = list(centrality.values())
node_size = [v * 10000 for v in cent_vals ]

nx.draw(g, 
        node_size=node_size,
        node_color=cent_vals,
        edge_color='white')
fig.set_facecolor('SaddleBrown')
```
